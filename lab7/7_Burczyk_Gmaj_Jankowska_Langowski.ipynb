## Implementacja atraktorow

a)

```python
def w1a(x, y): return( [0.5*x - 0.5*y, 0.5*x + 0.5*y])
def w2a(x, y): return([0.5*x + 0.5*y + 0.5, -0.5*x + 0.5*y + 0.5])
```

b)

```python
def w1b(x, y): return([0.5*x, 0.5*y])
def w2b(x, y): return([0.5*(x + 1), 0.5*y])
def w3b(x, y): return([0.25*(2*x + 1), 0.25*(2*y + np.sqrt(3))])
```

## Rysowanie atrakotora

### Stwórz dwie n-elementowe listy/tablice/krotki współrzędnych x oraz y.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10000
```

```python
x_a = [None]*n
y_a = [None]*n
x_b = [None]*n
y_b = [None]*n
```

```python
x_a[0] = np.random.uniform(high=0.99999)
y_a[0] = np.random.uniform(high=0.99999)
x_b[0] = np.random.uniform(high=0.99999)
y_b[0] = np.random.uniform(high=0.99999)
```

a) podział [0; 0.5) i [0.5; 1)

```python
def iteracje_a(x, y, n):
    for i in range(1, n):
        rnd = np.random.uniform(high=0.99999)
        if(rnd < 0.5):
            x[i] = w1a(x[i-1], y[i-1])[0]
            y[i] = w1a(x[i-1], y[i-1])[1]
        elif(rnd >= 0.5):
            x[i] = w2a(x[i-1], y[i-1])[0]
            y[i] = w2a(x[i-1], y[i-1])[1]
    
    return({'x': x, 'y': y})    
```

b) podział [0; 0.333333), [0.33333; 0.6666) i [0.6666, 1)

```python
def iteracje_b(x, y, n):
    for i in range(1, n):
        rnd = np.random.uniform(high=0.99999)
        if(rnd < 1/3):
            x[i] = w1b(x[i-1], y[i-1])[0]
            y[i] = w1b(x[i-1], y[i-1])[1]
        elif(rnd >=1/3 and rnd < 2/3):
            x[i] = w2b(x[i-1], y[i-1])[0]
            y[i] = w2b(x[i-1], y[i-1])[1]
        elif(rnd >= 2/3):
            x[i] = w3b(x[i-1], y[i-1])[0]
            y[i] = w3b(x[i-1], y[i-1])[1]                                         
    return({'x': x, 'y': y})    
```

dokonujemy losowania

```python
a = iteracje_a(x_a, y_a, n)
```

```python
b = iteracje_b(x_b, y_b, n)
```
